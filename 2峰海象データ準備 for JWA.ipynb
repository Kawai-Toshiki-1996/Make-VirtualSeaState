```python
freq=np.array([0.0445953,0.0486315,0.053033,0.0578329,0.0630672,0.0687753,0.075,0.0817881,0.0891906,0.097263,0.106066,0.1156658,0.1261345,0.1375506,0.15,0.1635762,0.1783811,0.194526,0.2121321,0.2313317,0.252269,0.2751013,0.3000001,0.3271524,0.3567623])
```

```python
def output(lambda_1,theta_1,s_1,omega_1, Hv_1,lambda_2,theta_2,s_2,omega_2,Hv_2,f):
    f.write(str('%03.8f' %(Hv_1))+","+str('%03.8f' %(math.cos(theta_1*math.pi/180)))+","+str('%03.8f' %(math.sin(theta_1*math.pi/180)))+","+str('%03.8f' %(lambda_1))+","+str('%03.8f' %(omega_1))+",")
    f.write(str('%03.8f' %(Hv_2))+","+str('%03.8f' %(math.cos(theta_2*math.pi/180)))+","+str('%03.8f' %(math.sin(theta_2*math.pi/180)))+","+str('%03.8f' %(lambda_2))+","+str('%03.8f' %(omega_2)))
    f.write('\n')

def def_S_max(H,omega_m):
    HL=H/(1.56*4*math.pi**2/omega_m**2)
    """波集中度s1の周波数依存"""
    if HL<0.026:
        #print((HL))
        log_10=0.4444*math.log10(HL)+0.5849
        a=10**(log_10)
        S_max=12.99*a**(-2.5)
        
    elif HL>=0.026:
        log10_gF=(0.1507-math.sqrt(0.1507**2+4*0.005882*0.8789+4*0.005882*math.log10(HL)))/(2*0.005882)
        gF=10**log10_gF
        gT=1.37*(1-1/(1+0.008*gF**(1/3))**5)
        S_max=12.99*gT**2.5
    
    if S_max >=84:
        S_max=84
    return S_max

def get_parameter_from_ind():
    """波高"""
    Hv_2  =10*(1-random.random())
    Hv_1  =10*(1-random.random())
    
    """尖度"""
    lambda_1 =10*(1-random.random())
    lambda_2 =10*(1-random.random())
    
    """波周波数"""
    T1=2.0+(15.0-2.0)*random.random()
    omega_1  =2.0*np.pi/T1#
    T2=2.0+(15.0-2.0)*random.random()
    omega_2  =2.0*np.pi/T2#0.3+1.7*random.random()
    
    """波向"""
    theta_1  =random.random()*360
    theta_2  =theta_1+5+350*random.uniform(0,1)
     
    """s"""
    s_1      =def_S_max(Hv_1,omega_1)#1+83*random.random()
    s_2      =def_S_max(Hv_2,omega_2)#1+83*random.random()
    
    
    return lambda_1,theta_1,s_1,omega_1, Hv_1,lambda_2,theta_2,s_2,omega_2,Hv_2


def spectrum_by_angle(lambda_1,theta_1,omega_1, Hv_1,lambda_2,theta_2,omega_2,Hv_2,omega,encount_angle,S_max1,S_max2):
 
    if omega<=omega_1:
        s_1=(S_max1*omega**5/omega_1**5)
    else:
        s_1=(S_max1*omega**(-2.5)/omega_1**(-2.5))
        
    spectrum_11=0.250*((4.0*lambda_1+1.0)/4.0*omega_1**4.0)**lambda_1
    spectrum_12=math.gamma(lambda_1)
    spectrum_16=Hv_1**2.0
    spectrum_13=omega**(4.0*lambda_1+1.0)
    spectrum_14=2.0**(2.0*s_1-1.0)*(math.gamma(s_1+1.0))**2.0/math.pi/math.gamma(2.0*s_1+1.0)
    spectrum_15=abs((math.cos((encount_angle-theta_1)/2.0*math.pi/180.0)))**(2.0*s_1)*math.exp(-(4.0*lambda_1+1.0)/4.0*(omega_1/omega)**4.0)
    spectrum_1=spectrum_11/spectrum_13*spectrum_14*spectrum_15*spectrum_16/spectrum_12

    
    if omega<=omega_2:
        s_2=(S_max2*omega**5/omega_2**5)
    else:
        s_2=(S_max2*omega**(-2.5)/omega_2**(-2.5))
        
    spectrum_21=0.250*((4.0*lambda_2+1.0)/4.0*omega_2**4.0)**lambda_2
    spectrum_22=math.gamma(lambda_2)
    spectrum_26=Hv_2**2.0
    spectrum_23=omega**(4.0*lambda_2+1.0)
    spectrum_24=2.0**(2.0*s_2-1.0)*(math.gamma(s_2+1.0))**2.0/math.pi/math.gamma(2.0*s_2+1.0)
    spectrum_25=abs((math.cos((encount_angle-theta_2)/2.0*math.pi/180.0)))**(2.0*s_2)*math.exp(-(4.0*lambda_2+1.0)/4.0*(omega_2/omega)**4.0)
    spectrum_2=spectrum_21/spectrum_23*spectrum_24*spectrum_25*spectrum_26/spectrum_22  
    
    return (spectrum_1+spectrum_2)*2*math.pi 
```

```python
for k in range(40000):
    if (k+1) % 100 ==0:
        print(k+1)
        
    random.seed()
    lambda_1,theta_1,s_1,omega_1, Hv_1,lambda_2,theta_2,s_2,omega_2,Hv_2 = get_parameter_from_ind()
    for i in freq:
        for j in range(36):
            f.write(str('%03.8f' %(spectrum_by_angle(lambda_1,theta_1,omega_1, Hv_1,lambda_2,theta_2,omega_2,Hv_2,i*np.pi*2,j*10,s_1,s_2)))+",")
    if omega_1 <= omega_2:
        output(lambda_1,theta_1,s_1,omega_1, Hv_1,lambda_2,theta_2,s_2,omega_2,Hv_2,f)
    else:
        output(lambda_2,theta_2,s_2,omega_2, Hv_2,lambda_1,theta_1,s_1,omega_1,Hv_1,f)

f.close()
print("end")
```
